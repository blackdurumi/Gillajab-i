```python
def mk_data_format(name:str):
    path='/content/drive/MyDrive/Voice_Data/KoG2P/txt_dataframe'
    
    os.mkdir(path+'/training_{}'.format(name))

    jamo_df = pd.read_csv(path+'/{}_jamo.csv'.format(name))

    path_list = list(jamo_df.path) 
    txt_iist = list(jamo_df.txt)
    name_list = [i.split('/')[-1].split('.')[0] for i in jamo_df.path]

    os.chdir(path+'/training_{}'.format(name))

    jamo_df.path = name_list  
    jamo_df.columns=['name','txt']
    jamo_df.to_csv('trainig_{}_text.csv'.format(name), index=False)

    jamo_df.txt = path_list
    jamo_df.columns=['name','path']
    jamo_df.to_csv('training_{}_wavscp.csv'.format(name), index=False)

    spk = [i.split('/')[1].split('.')[1][0]+'_'+i.split('/')[-1].split('.')[0][0]+'_'+i.split('/')[-1].split('.')[0].split('_')[1] for i in jamo_df.path]
    jamo_df.path = spk
    jamo_df.columns=['name', 'spk']
    jamo_df.to_csv('training_{}_utt2spk'.format(name), index=False)

    jamo_df=jamo_df[['spk', 'name']]
    jamo_df.to_csv('training_{}_spk2utt'.format(name), index=False)

```
