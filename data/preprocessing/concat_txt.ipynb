```python
cate_name = ['broadcast','hobby','dialog', 'life', 'weather', 'economy', 'play', 'shopping']
```

```python
def train_concat_df(cate_name):
    path = '/content/drive/MyDrive/Voice_Data/KoG2P/txt_dataframe/'  

    text_all_df = []
    utt2spk_all_df=[]
    spk2utt_all_df=[]
    null_path_all_df = []
    
    for i in tqdm_notebook(cate_name):
        text = pd.read_csv(path+'training_{}/trainig_{}_text.csv'.format(i,i))
        utt2spk = pd.read_csv(path+'training_{}/training_{}_utt2spk'.format(i,i))
        spk2utt = pd.read_csv(path+'training_{}/training_{}_spk2utt'.format(i,i))
        null_path = pd.read_csv(path+'training_{}/{}_null_path.csv'.format(i,i))


        text_all_df.append(text)
        utt2spk_all_df.append(utt2spk)
        spk2utt_all_df.append(spk2utt) 
        null_path_all_df.append(null_path)

    print('complete 1')     

    all_text = pd.concat(text_all_df,axis=0).reset_index(drop=True)
    all_utt2spk = pd.concat(utt2spk_all_df, axis=0).reset_index(drop=True)
    all_spk2utt = pd.concat(spk2utt_all_df,axis=0).reset_index(drop=True)
    all_null_path = pd.concat(null_path_all_df, axis=0).reset_index(drop=True)
    all_null_path.columns=['name']
    all_null_path.name = [i.split('/')[-1].split('.')[0] for i in all_null_path.name] 


    train_null_idx = list(all_text[all_text.txt.isna()].index)
    add_null = pd.DataFrame(list(all_text[all_text.txt.isna()].name))
    add_null.columns=['name']

    all_null = pd.concat((all_null_path, add_null), axis=0).reset_index(drop=True)

    all_text = all_text.drop(train_null_idx, axis=0).reset_index(drop=True)
    all_utt2spk = all_utt2spk.drop(train_null_idx, axis=0).reset_index(drop=True)
    all_spk2utt = all_spk2utt.drop(train_null_idx, axis=0).reset_index(drop=True)

    print('complete 2')

    all_text.to_csv(path+'concat/training_all_text.csv',index=False)
    all_utt2spk.to_csv(path+'concat/training_all_utt2spk.csv',index=False)
    all_spk2utt.to_csv(path+'concat/training_all_spk2utt.csv',index=False)
    all_null.to_csv(path+'concat/training_all_null_path.csv', index=False)
    
    print('complete 3')

    return all_text, all_utt2spk, all_spk2utt, all_null
```

```python
all_text, all_utt2spk, all_spk2utt, all_null_path = train_concat_df(cate_name)
```

```python
def val_concat_df(cate_name):
    path = '/content/drive/MyDrive/Voice_Data/KoG2P/txt_dataframe/'  

    text_all_df = []
    utt2spk_all_df=[]
    spk2utt_all_df=[]
    null_path_all_df = []
    
    for i in tqdm_notebook(cate_name):
        text = pd.read_csv(path+'val_{}/val_{}_text.csv'.format(i,i))
        utt2spk = pd.read_csv(path+'val_{}/val_{}_utt2spk.csv'.format(i,i))
        spk2utt = pd.read_csv(path+'val_{}/val_{}_spk2utt.csv'.format(i,i))
        try:
            null_path = pd.read_csv(path+'val_{}/val_{}_null_path.csv'.format(i,i))
        except EmptyDataError:
            print(f"empty {i} null path")
            pass

        text_all_df.append(text)
        utt2spk_all_df.append(utt2spk)
        spk2utt_all_df.append(spk2utt) 
        null_path_all_df.append(null_path)
        #print(f"complete {i}")

    print('complete 1')     

    all_text = pd.concat(text_all_df,axis=0).reset_index(drop=True)
    all_utt2spk = pd.concat(utt2spk_all_df, axis=0).reset_index(drop=True)
    all_spk2utt = pd.concat(spk2utt_all_df,axis=0).reset_index(drop=True)
    all_null_path = pd.concat(null_path_all_df, axis=0).reset_index(drop=True)
    all_null_path.columns=['name']
    all_null_path.name = [i.split('/')[-1].split('.')[0] for i in all_null_path.name] 


    val_null_idx = list(all_text[all_text.txt.isna()].index)
    add_null = pd.DataFrame(list(all_text[all_text.txt.isna()].name))
    add_null.columns=['name']

    all_null = pd.concat((all_null_path, add_null), axis=0).reset_index(drop=True)

    all_text = all_text.drop(val_null_idx, axis=0).reset_index(drop=True)
    all_utt2spk = all_utt2spk.drop(val_null_idx, axis=0).reset_index(drop=True)
    all_spk2utt = all_spk2utt.drop(val_null_idx, axis=0).reset_index(drop=True)

    all_spk2utt.drop('Unnamed: 0', axis=1, inplace=True)

    print('complete 2')

    all_text.to_csv(path+'concat/val_all_text.csv',index=False)
    all_utt2spk.to_csv(path+'concat/val_all_utt2spk.csv',index=False)
    all_spk2utt.to_csv(path+'concat/val_all_spk2utt.csv',index=False)
    all_null.to_csv(path+'concat/val_all_null_path.csv', index=False)
    
    print('complete 3')

    return all_text, all_utt2spk, all_spk2utt, all_null
```

```python
val_all_text, val_all_utt2spk, val_all_spk2utt, val_all_null_path = val_concat_df(cate_name)
```
